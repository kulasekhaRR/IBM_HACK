```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('2015_16_Statewise_Elementary.csv')
print(df.head())
state_avg_literacy = df.groupby('STATNAME')['OVERALL_LI'].mean().sort_values(ascending=True)
plt.figure(figsize=(12, 6))
state_avg_literacy.plot(kind='bar', color='red')
plt.title('States with the Lowest Literacy Rates')
plt.xlabel('State')
plt.ylabel('Average Literacy Rate')
plt.show()




grouped_data = df.groupby('STATNAME')[['FEMALE_LIT', 'MALE_LIT']].sum()
grouped_data['GENDER_RATIO'] = (grouped_data['FEMALE_LIT'] / grouped_data['MALE_LIT']) * 100
plt.figure(figsize=(12, 6))
grouped_data['GENDER_RATIO'].sort_values().plot(kind='bar', color='skyblue')
plt.title('Gender Ratio of Literates by State')
plt.xlabel('State')
plt.ylabel('Gender Ratio (%)')
plt.xticks(rotation=90)
plt.show()




grouped_data = df.groupby('STATNAME')[['FEMALE_LIT', 'MALE_LIT']].sum()
grouped_data['GENDER_RATIO'] = (grouped_data['FEMALE_LIT'] / grouped_data['MALE_LIT']) * 100
plt.figure(figsize=(8, 8))
plt.pie(grouped_data['GENDER_RATIO'], labels=grouped_data.index, autopct='%1.1f%%', startangle=140)
plt.title('Gender Ratio of Literates by State')
plt.axis('equal')

plt.show()




grouped_data = df.groupby('STATNAME')[['FEMALE_LIT', 'MALE_LIT']].sum()
grouped_data['GENDER_DIFF'] = abs(grouped_data['FEMALE_LIT'] - grouped_data['MALE_LIT'])
plt.figure(figsize=(12, 6))
grouped_data['GENDER_DIFF'].sort_values().plot(kind='bar', color='purple')
plt.title('Difference Between Male and Female Literates by State')
plt.xlabel('State')
plt.ylabel('Gender Difference')
plt.xticks(rotation=90)
plt.show()




grouped_data = df.groupby('DISTRICTS')[['FEMALE_LIT', 'MALE_LIT']].sum()
grouped_data['GENDER_DIFF'] = abs(grouped_data['FEMALE_LIT'] - grouped_data['MALE_LIT'])
grouped_data = grouped_data.sort_values(by='GENDER_DIFF', ascending=False)
plt.figure(figsize=(10, 12))
plt.barh(grouped_data.index, grouped_data['GENDER_DIFF'], color='green')
plt.title('Difference Between Male and Female Literates by District')
plt.xlabel('Gender Difference')
plt.ylabel('District')
for i, value in enumerate(grouped_data['GENDER_DIFF']):
    plt.text(value, i, f' {grouped_data.index[i]} ({value})', va='center')

plt.show()




grouped_data = df.groupby(['STATNAME', 'DISTRICTS', 'VILLAGES'])[['FEMALE_LIT', 'MALE_LIT']].sum()
grouped_data['GENDER_DIFF'] = abs(grouped_data['FEMALE_LIT'] - grouped_data['MALE_LIT'])
district_data = grouped_data.groupby(['STATNAME', 'DISTRICTS'])['GENDER_DIFF'].sum()
plt.figure(figsize=(10, 12))
plt.barh([f'{name[0]} - {name[1]}' for name in district_data.index], district_data.values, color='green')
plt.title('Difference Between Male and Female Literates by District')
plt.xlabel('Gender Difference')
plt.ylabel('District')
plt.show()



grouped_data = df.groupby(['STATNAME', 'DISTRICTS'])[['FEMALE_LIT', 'MALE_LIT']].sum()
grouped_data['GENDER_DIFF'] = abs(grouped_data['FEMALE_LIT'] - grouped_data['MALE_LIT'])
plt.figure(figsize=(10, 12))
plt.barh([f'{name[0]} - {name[1]}' for name in grouped_data.index], grouped_data['GENDER_DIFF'], color='green')
plt.title('Difference Between Male and Female Literates by District')
plt.xlabel('Gender Difference')
plt.ylabel('District')

plt.show()



state_district_stats = df.groupby(['STATNAME', 'DISTRICTS']).agg({
    'OVERALL_LI': 'mean',
    'POPULATION_0_6': 'mean',
    'FEMALE_LIT': 'mean',
    'MALE_LIT': 'mean'
}).reset_index()
lowest_overall_li = state_district_stats[state_district_stats['OVERALL_LI'] == state_district_stats['OVERALL_LI'].min()]
highest_pop_below_6 = state_district_stats[state_district_stats['POPULATION_0_6'] == state_district_stats['POPULATION_0_6'].max()]
print("State and District with the Lowest Overall Literacy Rate:")
print(lowest_overall_li)
print("\nState and District with the Highest Percentage of Population Below 6 Years:")
print(highest_pop_below_6)



import seaborn as sns
selected_columns = ['OVERALL_LI', 'POPULATION_0_6', 'FEMALE_LIT', 'MALE_LIT']
correlation_matrix = df[selected_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap for Education Data')
plt.show()
```
